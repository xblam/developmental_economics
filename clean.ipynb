```python
import pandas as pd
import re

# Read CSV
df = pd.read_csv("data/data.csv")

# Inspect
print(list(df.columns))
print(len(df.columns))
```

```python
yinc_cols = [c for c in df.columns if c.startswith("YINC-")]

(df[yinc_cols] > 0).sum()
```

```python
keep_exact = [
    "PUBID",
    "KEY!SEX",
    "KEY!BDATE_Y",
    "KEY!RACE_ETHNICITY",
]

keep_prefixes = [
    "CV_MSA",          # location / treatment
    "CV_CENSUS_REGION",
    "CV_URBAN-RURAL",
    "CV_HH",
    "YINC-1700",       # main outcome
    "YINC-1800",       # robustness outcome
]

def keep_col(col):
    if col in keep_exact:
        return True
    return any(col.startswith(prefix) for prefix in keep_prefixes)

cols_to_keep = [c for c in df.columns if keep_col(c)]
```

```python
df_clean = df[cols_to_keep].copy()
df_clean = df_clean.sort_index(axis=1)
df_clean.to_csv("data/nlsy97_msa_migration_panel.csv", index=False)

```

```python
df_clean.info()
print(list(df_clean.columns))
```

```python
year_map = {
    0: 1997,
    1: 1998,
    2: 1999,
    3: 2000,
    4: 2001,
    5: 2002,
    6: 2003,
    7: 2004,
    8: 2005,
    9: 2006,
    10: 2007,
    11: 2008,
    12: 2009,
    13: 2010,
    14: 2011,
    15: 2013,
    16: 2015,
    17: 2017,
    18: 2019,
    19: 2021
}

def rename_col(col):
    # identifiers (no year)
    if col in ["PUBID", "KEY!SEX", "KEY!BDATE_Y", "KEY!RACE_ETHNICITY"]:
        return col.lower()

    # match variable + optional year suffix
    m = re.match(r"(.+?)(?:\.(\d+))?$", col)
    base, idx = m.groups()

    year = year_map[int(idx)] if idx is not None else 1997

    base_map = {
        "CV_MSA": "in_msa",
        "CV_CENSUS_REGION": "census_region",
        "CV_URBAN-RURAL": "urban_rural",
        "CV_HH_SIZE": "hh_size",
        "YINC-1700": "income_wages",
        "YINC-1800": "income_wages_est",
    }

    if base not in base_map:
        return col.lower()

    return f"{base_map[base]}_{year}"
```

```python
df2 = df_clean.rename(columns={c: rename_col(c) for c in df.columns})
df2.to_csv("data/df2.csv", index=False)
```

```python
print(list(df2.columns))
```

```python
id_vars = [
    "pubid",
    "key!sex",
    "key!race_ethnicity",
    "key!bdate_y",
    "cv_hh_income_source"
]
long_df = (
    df2
    .melt(
        id_vars=id_vars,
        var_name="var_year",
        value_name="value"
    )
    .dropna(subset=["value"])
)
# Extract year (last 4 digits)
long_df["year"] = long_df["var_year"].str.extract(r"_(\d{4})").astype(int)

# Extract variable name (everything before _YYYY)
long_df["variable"] = long_df["var_year"].str.replace(r"_\d{4}$", "", regex=True)
panel_df = (
    long_df
    .pivot_table(
        index=id_vars + ["year"],
        columns="variable",
        values="value"
    )
    .reset_index()
)
panel_df.columns.name = None
panel_df = panel_df.sort_values(["pubid", "year"]).reset_index(drop=True)
panel_df["age"] = panel_df["year"] - panel_df["key!bdate_y"]
panel_df = panel_df[(panel_df["age"] >= 13) & (panel_df["age"] <= 18)]
```
